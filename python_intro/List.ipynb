## slicing
* it slices the given list according the position we provide and returns a new list rather than slicing the original list
* formula =[ start_index : stop_index]
* parameters
* start index : slicing starts at given position including that position
* if start position is not provided by default slicing start at 0 index


```python
num=[100,200,300,400,500,600,700]
num[1:]
```

```python
num[1:5]
```

```python
num[:5]
```

```python
mix=[110,200,324,"Apple","Mango",32,67.98,120]
# now slice the list strting from apple to end
```

```python
mix[3:]
```

```python
mix[:]
```

## stop index 
* Index position where we stop the slicing
* but unlike start the stop excludes the stop positioned slicice till previous element
* if start index is 1:4 slicing start at 1 but stops at index 3 excluding index 4
